```python
def get_skill_gaps(student_row, career):

    reqs = career_requirements[career]

    gaps = []

    strengths = []

    for skill, target in reqs.items():

        current = float(student_row.iloc[0][skill])

        if current >= target:

            strengths.append(
                (skill, current)
            )

        else:

            gaps.append({
                "skill": skill,
                "current": current,
                "target": target,
                "gap": round(target-current, 1)
            })

    gaps = sorted(
        gaps,
        key=lambda x: x["gap"],
        reverse=True
    )

    return strengths, gaps
```
