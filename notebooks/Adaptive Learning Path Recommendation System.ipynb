```python
import torch
from torch.utils.data import Dataset ,DataLoader

class AdaptiveLearningDataset(Dataset):

    def __init__(self, X, y_diff, y_topic):

        self.X = torch.tensor(X, dtype=torch.float32)

        self.y_diff = torch.tensor(
            y_diff,
            dtype=torch.long
        )

        self.y_topic = torch.tensor(
            y_topic,
            dtype=torch.long
        )

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):

        return (
            self.X[idx],
            self.y_diff[idx],
            self.y_topic[idx]
        )
```

Creating Dataset Objects

```python
train_dataset=AdaptiveLearningDataset(X_train, y_diff_train,y_topic_train)
test_dataset=AdaptiveLearningDataset(X_test,y_diff_test,y_topic_test)
```

making DataLoaders

```python
train_loader = DataLoader(train_dataset, batch_size=256, shuffle=True)

test_loader = DataLoader(test_dataset,batch_size=256)
```

Making the embedded model

```python
import torch.nn as nn

class AdaptiveLearningModel(nn.Module):

    def __init__( self,num_features, num_topics):

        super().__init__()

        self.embedding = nn.Sequential(

            nn.Linear(num_features, 128),
            nn.ReLU(),

            nn.BatchNorm1d(128),
            nn.Dropout(0.3),

            nn.Linear(128, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU()
            )

        self.difficulty_head = nn.Linear(32, 3)

        self.topic_head = nn.Linear(32,num_topics)

    def forward(self, x):

        emb = self.embedding(x)

        difficulty_logits = self.difficulty_head(emb)

        topic_logits = self.topic_head(emb)

        return (difficulty_logits, topic_logits, emb)
```

```python
num_features = X_train.shape[1]

num_topics = len( topic_encoder.classes_)

model = AdaptiveLearningModel( num_features, num_topics)
```

```python
completed_topics = [
    "Python Basics",
    "NumPy"
]
```

```python
def get_available_topics(
    completed,
    graph
):

    available = []

    for node in graph.nodes():

        if node in completed:
            continue

        prereqs = list(
            graph.predecessors(node)
        )

        if all(
            p in completed
            for p in prereqs
        ):
            available.append(node)

    return available
```

```python
available = get_available_topics(
    completed_topics,
    G
)

print(available)
```

```python
eligible = get_available_topics(
    completed_topics,
    G
)

filtered = {
    k:v
    for k,v in recommendation_scores.items()
    if k in eligible
}
```

```python
top_topics = sorted(
    filtered.items(),
    key=lambda x:x[1],
    reverse=True
)

print(top_topics)
```

```python
def generate_roadmap(
    completed,
    graph
):

    roadmap = []

    completed = set(completed)

    while True:

        available = get_available_topics(
            completed,
            graph
        )

        if not available:
            break

        roadmap.extend(
            available
        )

        completed.update(
            available
        )

    return roadmap
```

```python
roadmap = generate_roadmap(
    completed_topics,
    G
)

print(roadmap)
```
