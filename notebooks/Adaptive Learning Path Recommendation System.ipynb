```python
import torch
from torch.utils.data import Dataset

class AdaptiveLearningDataset(Dataset):

    def __init__(self, X, y_diff, y_topic):

        self.X = torch.tensor(X, dtype=torch.float32)

        self.y_diff = torch.tensor(
            y_diff,
            dtype=torch.long
        )

        self.y_topic = torch.tensor(
            y_topic,
            dtype=torch.long
        )

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):

        return (
            self.X[idx],
            self.y_diff[idx],
            self.y_topic[idx]
        )
```
