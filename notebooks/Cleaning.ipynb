```python
num_cols = [
    'avg_session_duration',
    'cloud_score',
    'quiz_avg_score',
]

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
df["streak_days"] = df["streak_days"].fillna(0)
```

```python
cat_cols = [
    'career_goal',
    'recommended_next_topic'
]

for col in cat_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
df.to_csv(
    "student_profiles_clean.csv",
    index=False
)
```
